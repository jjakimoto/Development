```python
import numpy as np
from random import gauss
from itertools import product


def batch(coeffs, n_iter=1e5, max_hp=100,
          pts=np.linspace(.5, 10, 20),
          sls=np.linspace(.5, 10, 20),
          seed=0):
    n_iter = int(n_iter)
    phi = 2 ** (-1. / coeffs['hl'])
    results = []
    for pt_sl in product(pts, sls):
        output = []
        for _ in range(n_iter):
            p = seed
            hp = 0
            count = 0
            while True:
                p = (1 - phi) * coeffs['forecast']\
                    + phi * p + coeffs['sigma'] * gauss(0, 1)
                diff = p - seed
                hp += 1
                if diff > pt_sl[0] or diff < pt_sl[1] or hp > max_hp:
                    output.append(diff)
                    break
        mean = np.mean(output)
        std = np.std(output)
        print(pt_sl[0], pt_sl[1], mean, std, mean/std)
        results.append((pt_sl[0], pt_sl[1], mean, std, mean/std))
    return results
```

```python
from tqdm import tqdm_notebook

pts = np.linspace(0, 10, 21)
sls = np.linspace(0, 10, 21)
count = 0
hls = [5,10,25,50,100]
forecasts = [10,5,0,-5,-10]
outputs = []
for prod_ in tqdm_notebook(product(forecasts, hls)):
    count += 1
    coeffs = {'forecast': prod_[0], 'hl': prod_[1], 'sigma': 1}
    output = batch(coeffs, n_iter=1e5, max_hp=100, pts=pts, sls=sls)
    outputs.append((coeffs, output))
```
